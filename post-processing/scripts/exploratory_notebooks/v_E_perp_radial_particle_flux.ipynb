```python
# Imports.
import numpy as np;
import xarray as xr;
import matplotlib.pyplot as plt;
import h5_reader_xr as reader;
import gysela_utilities as gys_utils;
import rprof_utilities as utils;

# Styling.
plt.style.use("ggplot");
```

```python
# Test inputs - delete when done!
directory_path = "/zhisongqu_data/seth/GYSELA/raw/fresh_batch_10.0/DN_FLUXDRIVEN_ITG_TEST";
data_arrays = reader.fetch_rprof_data(directory_path, dataset = "GammaGC_vE_r_FSavg", parallelise = True);
```

```python
def plot_radial_particle_flux_time_series(rprof_list, dt_diag, effective_radius = None, log_scale = False):	
	# Stack into (time, r) array.
	particle_flux_series = xr.concat(rprof_list, dim = "time");
	
	if effective_radius is not None:
		particle_flux_series = gys_utils.slice_at_effective_radius(particle_flux_series, effective_radius);
	
	time_range = gys_utils.generate_time_range_by_series(particle_flux_series, dt_diag);

	# Figure plotting logic.
	operation = plt.semilogy if log_scale else plt.plot;
	rho_label = f"$\\rho = {effective_radius}$" if effective_radius is not None else "all radii";
	plt.figure(figsize=(10, 5));
	operation(time_range, particle_flux_series, color = "blue", lw = 1.0, label = r"$\Gamma_{E \times B}$");
	plt.title(rf"Radial $E \times B$ Particle Flux, {rho_label}", fontsize = 14);
	plt.xlabel("Time [GYSELA code units]", fontsize = 12);
	plt.ylabel(r"$\Gamma_{E \times B}$", fontsize = 12);
	plt.legend();
	plt.show();

plot_radial_particle_flux_time_series(data_arrays, dt_diag = 250, effective_radius = 0.7, log_scale = False);
```
